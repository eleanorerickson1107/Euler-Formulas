```python
#Block 4
#Forward Euler's Method, Heun's Method, and Runge-Kutta Method functions, don't touch, just run!

def Euler(h,n,t_var,y_var):
    #n=int(n*h**(-1)-(n*h**(-1)))
    #n=n-1
    mat=np.empty([t_var,y_var])
    default_row=[t_var,y_var]
    mat = np.append(mat,[default_row], axis=0)
    for i in range(n):
        try:
            y_d=y_derive(t_var,y_var)
            y_var=y_var+h*y_d
            new_row=[t_var+h,"{:f}".format(y_var)]
            mat = np.append(mat, [new_row], axis=0)
            new_row.clear()
            t_var=t_var+h
        except:
            new_row=[t_var+h,"Overflow"]
            mat = np.append(mat, [new_row], axis=0)
            new_row.clear()
            t_var=t_var+h
    return mat
def Heun(h,n,t_var,y_var):
    mat=np.empty([t_var,y_var])
    default_row=[t_var,y_var]
    mat = np.append(mat,[default_row], axis=0)
    for i in range(n):
        try:
            y_d=y_derive(t_var,y_var)
            t_n=t_var+h
            y_n=y_var+h*y_d
            y_var=y_var+0.5*h*y_derive(t_n,y_n)
            new_row=[t_var+h,"{:f}".format(y_var)]
            mat = np.append(mat, [new_row], axis=0)
            new_row.clear()
            t_var=t_var+h
        except:
            new_row=[t_var+h,"Overflow"]
            mat = np.append(mat, [new_row], axis=0)
            new_row.clear()
            t_var=t_var+h
    return mat
def RungeKutta(h,n,t_var,y_var):
    mat=np.empty([t_var,y_var])
    default_row=[t_var,y_var]
    mat = np.append(mat,[default_row], axis=0)
    for i in range(n):
        try:
            kn1=y_derive(t_var,y_var)
            kn2t=t_var+0.5*h
            kn2y=y_var+0.5*h*kn1
            kn2=y_derive(kn2t,kn2y)
            kn3y=y_var+0.5*h*kn2
            kn3=y_derive(kn2t,kn3y)
            kn4t=t_var+h
            kn4y=y_var+h*kn3
            kn4=y_derive(kn4t,kn4y)
            y_var=y_var+h*((kn1+2*kn2+2*kn3+kn4)/6)
            new_row=[t_var+h,"{:f}".format(y_var)]
            mat = np.append(mat, [new_row], axis=0)
            new_row.clear()
            t_var=t_var+h
        except:
            new_row=[t_var+h,"Overflow"]
            mat = np.append(mat, [new_row], axis=0)
            new_row.clear()
            t_var=t_var+h
    return mat
```
